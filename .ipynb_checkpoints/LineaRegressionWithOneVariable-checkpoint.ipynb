```octave
x = [ones(m,1), dt(:,1)];
iterations = 1500;
alpha = 0.02;
theta = zeros(2,1);
Jhistory = zeros(1500,1);
```

```octave

function J = getCost(x,y,theta)
    m = length(x);
    hypothesis = x * theta;
    sqrError = (hypothesis - y).^2;
    sumOfSquare = sum(sqrError);

    J = (1/(2*m))*sumOfSquare;
endfunction


% Implementing Gradient Descent

for i=1:1500

theta = theta - (((x'*(x*theta-y))./m)*alpha);

endfor


```
